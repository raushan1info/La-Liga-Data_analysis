```python
# Useful library
import pandas as pd
```

```python
# reading csv, header parameter to avoid system generated column names
df=pd.read_csv('Laliga.csv',header=1)
# Top 5 records
df.head()
```

1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

```python
# Counting '-' value in each column

(df=='-').sum()
```

```python
# replacing dashes
df.replace('-',0,inplace=True)
```

```python
#after replacing, checking count of dashes
(df=='-').sum()
```

```python
# Converting object to int64 datatype to perform mathematical operations except 3 columns ('Since/LastApp','Debut','Team')
for col in df.columns:
  if col not in ['Since/LastApp','Debut','Team']:
    df[col]=df[col].astype('int64')

df.info()
```

2. Print all the teams which have started playing between 1930-1980

```python
df[df.Debut.between('1930','1980')][['Team','Debut']]
```

3. Print the list of teams which came Top 5 in terms of points

```python
df.nlargest(5,columns='Points')[['Team','Points']]
```

4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function,find the team which has maximum and minimum goal difference.

```python
def Goal_diff_count():
  df['Goal_Difference']=df['GoalsFor']-df['GoalsAgainst']
  return df[['Team','Goal_Difference']]

Goal_diff_count()
```

```python
print('The team which has maximum goal difference')
Goal_diff_count().nlargest(1,columns='Goal_Difference')
```

```python
print('The team which has minimum goal difference')
Goal_diff_count().nsmallest(1,columns='Goal_Difference')
```

5. Create a new column with name “Winning Percent” and append it to the data set.Percentage of Winning = (GamesWon / GamesPlayed)*100 If there are any numerical error, replace it with 0%. Print the top 5 teams which has the highest Winning percentage

```python
#Adding new column “Winning_Percent”
df['Winning_Percent']=(df.GamesWon/df.GamesPlayed)*100
```

```python
#checking Nan value in new column
df.Winning_Percent.isnull().sum()
```

```python
# Replace Nan value with 0%
df.Winning_Percent.fillna(0,inplace=True)
```

```python
# top 5 teams which has the highest Winning percentage
df[['Team','Winning_Percent']].nlargest(5,'Winning_Percent').round(1)
```
